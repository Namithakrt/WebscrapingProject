```python
def correct_email(email):
    # Replace spaces with nothing
    email = email.replace(" ", "")
    
    # Add '@' if missing (assumes a common domain)
    if "@" not in email:
        if "gmail.com" in email:
            email = email.replace("gmail.com", "@gmail.com")
        elif "yahoo.com" in email:
            email = email.replace("yahoo.com", "@yahoo.com")
        elif "hotmail.com" in email:
            email = email.replace("hotmail.com", "@hotmail.com")
        elif ".info" in email:
            email = email.replace(".info","@gmail.com")
        elif ".biz" in email:
            email = email.replace(".biz","@gmail.com")
        elif ".com" in email:
            email = email.replace(".com","@gmail.com")
        else:
            email+="@gmail.com"
            
    # Ensure domain has a dot
    if "@" in email:
        local_part, domain = email.split("@")
        if ".biz" in domain:
            domain = domain.replace(".biz",".com")
        elif ".info" in domain:
            domain = domain.replace(".info",".com")
        elif ".org" in domain:
            domain = domain.replace(".org",".com")
        elif ".net" in domain:
            domain = domain.replace(".net",".com")
        email = f"{local_part}@{domain}"
    
    return email

df2.loc[:,'Email'] = df2['Email'].apply(correct_email)
```

```python
#cleaning names

import re

def correct_name(name):
    # Remove titles using regular expressions
    name = re.sub(r'^(Mr\.|Mrs\.|Dr\.)\s*', '', name)
    # Split the name into parts
    parts = name.split()
    if len(parts) < 2:
        return name  # return the name as is if it doesn't have at least two parts
    first_name = parts[0]
    last_name = parts[1]
    # Remove any glued extraneous words from the last name
    pattern = r"\b\w+$"
    last_name = re.sub(pattern," ",last_name)
    cleaned_name = f"{first_name} {last_name}"
    return cleaned_name
# Apply the cleaning function to the Name column
df2.loc[:,'Name'] = df2['Name'].apply(correct_name)
```

```python
df2.loc[:,'Join Date'] = df2['Join Date'].str.replace('/', '-')

def convert_dates(date_str):
    try:
        # First, attempt to parse with day-first
        return pd.to_datetime(date_str, format='%d-%m-%Y').strftime('%Y-%m-%d')
    except ValueError:
        try:
            # If day-first fails, attempt to parse with year-first
            return pd.to_datetime(date_str, format='%Y-%m-%d').strftime('%Y-%m-%d')
        except ValueError:
            # If both parsing attempts fail, return None
            return None
df2.loc[:,'Join Date'] = df2['Join Date'].apply(convert_dates)
```

```python
job_titles = ['Engineering','Support','HR','Sales','Marketing']

def clean_job(dept):
    for title in job_titles:
        if dept.startswith(title):
            return title
    return dept

df2.loc[:,'Department'] = df2['Department'].apply(clean_job)
    
```
